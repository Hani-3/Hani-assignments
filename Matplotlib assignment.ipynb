Question 2

A scatter plot uses dots to represent values for two different nummeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y) # to plot scatter plot
plt.xlabel('x axis') # to add xlabel
plt.ylabel('y axis') # to add y label
plt.title('scatter plot') # to add title
plt.show() # to show plot
```

Question 3

The subplot function in Matplotlib is used to create multiple plots on a single figure. This can be useful for comparing different datasets or visualizing different aspects of the same data.

The subplot function takes three arguments: 
- number of raws
- number of columns
- index of the plot
The index starts at 1 in the upper left corner and goes across the top row to the right, then down a row and across again, and so on.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5]) # given data for x axis
y = np.array([0, 100, 200, 300, 400, 500]) # given data for y axis
plt.subplot(2,2,1) # to create subplot, here first 2=no of raws, second 2=no of columns, 1=index of figure
plt.plot(x,y) # to make first plot

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show() # to show plot
```

Question 4

A bar plot is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

Bar plots are useful for comparing different sets of data among different groups. They can also illustrate important changes in data throught a period  of time.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit) # to plot a bar plot
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.show()
```

```python
plt.barh(company,profit) # to plot a horizontal bar plot
plt.xlabel('Profit')
plt.ylabel('Company Name')
plt.show()
```

A box plot is a type of chart that shows the distribution of data. It is also known as a box-and-whisker plot. Box plots are used to visualize the five-number summary of a dataset: the minimum, first quartile, median, third quartile, and maximum.

The box in the plot represents the middle 50% of the data. The whiskers extend from the box to the minimum and maximum values. The median is represented by a line inside the box.

Box plots are useful for comparing the distributions of different datasets. They can also be used to identify outliers, which are data points that are significantly different from the rest of the data.

Box plots are useful for several purposes:
- Visualizing Data Distribution: Box plots provide a visual summary of the distribution of a dataset, including the center, spread, and skewness.
- Comparing Distributions: Box plots make it easy to compare the distribution of multiple datasets or categories.
- Identifying Outliers: The whiskers and outliers on a box plot can help identify data points that may be unusually high or low.
- Understanding Central Tendency: The median and quartiles displayed in the box plot provide insights into the central tendency of the data.

```python
box1 = np.random.normal(100, 10, 200) # given data of box 1
box2 = np.random.normal(90, 20, 200) # given data of box 2
box = [box1, box2] # make list of both datasets
plt.boxplot(box) # to plot both dataset in one figure
plt.show() # to show plot
```
