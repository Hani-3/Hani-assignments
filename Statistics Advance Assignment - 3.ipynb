**Q1: What is Estimation Statistics? Explain point estimate and interval estimate.**

Estimation statistics is a data analysis framework that uses a combination of techniques to plan experiments, analyze data, and interpret results. 
In statistics, estimation is a procedure used to calculate the value of a population property from observations of a sample drawn from the population. It is an observed numerical value used to estimate an unknown populating parameter.

There are two types of Estimates:
- Point Estimate: A single value chosen from a sample to estimate a population parameter. A point estimate is a sample statistic calculated using the sample data to estimate the most likely value of the corresponding unknown population parameter.
- Interval Estimates: A range of values within which the population parameter lies. Interval estimation involves constructing a range of values, called a confidence interval, that is likely to contain the unknown population parameter with a certain level of confidence. 

**Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.**

```python
import numpy as np
population = np.random.normal(50,10,1000) # create random population data
sample = np.random.choice(population, 10) # create sample from population data
print(f'Population mean is {np.mean(population)}') # actual population mean
print(f'population standard deviation is {np.std(population)}') # actual populaton std
print(f"Estimated population mean using sample mean is {np.mean(sample)}") # print mean of sample data
print(f"Estimated population standard deviation using sample standard deviation is {np.std(sample)}") # print std of sample data
```

**Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.**

Hypothesis testing is a statistical method that helps researchers and decision-makers make informed decisions based on evidence. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (H1), and then collecting data to assess the evidence. 

Hypothesis testing is important because it allows researchers to systematically gather evidence to support or refute a proposed explanation or hypothesis. It can also help professionals test their theories and assumptions before putting them into action.

**Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.**

```python
import scipy
population1 = np.random.randint(1,50,50) # create population1 data
sample1 = np.random.choice(population1,20) # create sample data from population1 data
population2 = np.random.randint(1,100,50)  # create population2 data
sample2 = np.random.choice(population2,20) # create sample data from population2 data
result = scipy.stats.ttest_ind(sample1,sample2) # do ttest on both sample
print(f'Mean of sample1 is {np.mean(sample1)}') # to print mean of sample 1
print(f'Mean of sample2 is {np.mean(sample2)}') # to print mean of sample 2
if result[1] <= 0.05: # make condition to reject or accept null hypothesis
    print('Reject the null hypothesis both sample mean are not same')
else:
    print('Accept the null hypothesis both sample mean are same')
```

**Q8. Define p-value and explain its significance in hypothesis testing.**

The p-value, or probability value, is a measure used in hypothesis testing to determine the strength of evidence against the null hypothesis. It represents the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed in the sample data, under the assumption that the null hypothesis is true.

Significance of the p-value in hypothesis testing is that it helps researchers make decisions about whether to reject or retain the null hypothesis based on the strength of evidence provided by the sample data. A lower p-value indicates stronger evidence against the null hypothesis, while a higher p-value suggests weaker evidence. However, it's important to remember that the p-value alone does not provide information about the size or practical importance of any observed effect; it only indicates the strength of evidence against the null hypothesis.

**Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom

**Q11: What is Student’s t distribution? When to use the t-Distribution.**

Student's t-distribution is a continuous probability distribution that is bell-shaped and symmetric around zero. It is similar to the normal distribution, but has heavier tails. The t-distribution is defined by the degrees of freedom, which are related to the sample size. As the sample size increases, the t-distribution becomes more similar to a normal distribution. 

Uses:
- Estimating population parameters for small sample sizes (n<30) or unknown variances
- Making presumptions about a mean when the standard deviation is not known
- Assessing the statistical significance of the difference between two sample means
- Constructing confidence intervals for the difference between two population means
- Linear regression analysis 

**Q12: What is t-statistic? State the formula for t-statistic.**

T-statistics is test statistics. test statistic measures the accuracy of the predicted data distribution relating to the null hypothesis you use when analyzing data samples. The statistic depends on what kind of data analysis method you use and indicates how closely your data matches the predicted distribution for the specific test you perform. The distribution of data accounts for the frequency at which observations occur when performing statistical tests and shows the central tendency and variation.

Formula:
$t = \frac{\bar{x} - \mu_{0}}{\frac{s}{\sqrt{n}}}$   
where,  
$\bar{x}$ is sample mean  
$\mu$ is population mean  
$s$ is sample standard deviation  
$n$ is sample size

**Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be \\$500 with a standard deviation of \\$50.
Estimate the population mean revenue with a 95% confidence interval.**

$\text{Confidence Interval} = \bar{x} ± {z} \times {\frac{s}{\sqrt{n}}}$  
where, 
$\bar{x}$ is sample mean  
$z$ is z-score corresponding to confidence level   
$s$ is sample standard deviation   
$n$ is sample size   
$\text{Confidence Interval} = 500 ± 1.96 \times {\frac{50}{\sqrt{50}}}$   
$= 500 ± 1.96 \times 7.07$   
$= 500 ± 13.857$   
$= 486.14 \text{ to } 513.86$   
The population mean revenue with a 95% confidence interval is in range of 486.14 to 513.86.

**Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.**

Null Hypothesis : drug decreases blood pressure by 10 mmHg. $(\mu < 10)$   
Alternative Hypothesis : drug does not decrease blood pressure by 10 mmHg. $(\mu=10)$   
significance level $(\alpha = 0.05)$  
degree of freedom = n - 1 = 100 - 1 = 99  
t-test:  
$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$  
$t = \frac{8 - 10}{\frac{3}{\sqrt{100}}}$  
$t = -6.667$  
the critical t-value for a significance level of 0.05 and 99 degrees of freedom is approximately 1.984.  
$-6.667<1.984$ so we accept the null hypothesis new drug will decrease blood pressure by 10 mmHg.

**Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.**

Null Hypothesis: Mean of product weight is 5 pounds $(\mu = 5)$  
Alternative Hypothesis: Mean of product weight is less then 5 pounds $(\mu < 5)$  
significance level $(\alpha =  0.01)$  
z-score:  
$z = \frac{\bar{x}-\mu}{\frac{\sigma}{\sqrt{n}}}$   
$z = \frac{4.8 - 5}{\frac{0.5}{\sqrt{25}}}$    
$z = -2$   
For a significance level of 0.01, the critical Z-value is approximately -2.33.
$-2>-2.33$ so, we reject the null hypothesis mean of weight is less than 5 pounds.

**Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.**

sample 1:  
sample size $n_{1}=30$  
sample mean $\bar{x_{1}} = 80$  
standard deviation $s_{1} = 10$  
sample 2:    
sample size $n_{2}=40$  
sample mean $\bar{x_{2}} = 75$  
standard deviation $s_{2} = 8$  

$H_{0} : \mu_{1} = \mu_{2}$  
$H_{1} : \mu_{1} \neq \mu_{2}$

Pooled standard deviation $s_{p}$:  
$s_{p} = \sqrt{\frac{(n_{1}-1)s_{1}^2 + (n_{2}-1)s_{2}^2}{n_{1}+n_{2}-2}}$  
$s_{p} = \sqrt{\frac{(30-1)10^2 + (40-1)8^2}{30+40-2}}$  
$s_{p} = 8.908$

test statistic t:  
$t = \frac{(\bar{x_{1}} - \bar{x_{2}})}{s_{p}(\sqrt{\frac{1}{n_{1}}+\frac{1}{n_{2}}})}$
$t = \frac{(80 - 75)}{8.908(\sqrt{\frac{1}{30}+\frac{1}{40}})}$  
$t = 2.3246$
p value = 0.02309  
since 0.02309 > 0.01 we fail to reject null hypothesis population means for two groups are equal.
