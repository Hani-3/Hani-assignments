The ArothmaticError class is a base class for exceptions that occur during arithmatic operations in Python. It itself doesn't define specific errors, but it serves as a base class for more specific arithmatic related exceptions. Here, I will take two noteble exceptions derived from ArithmaticError are ZeroDivisionError and OverflowError.

```python
# Example 1: ZeroDivisionError
try: #try block
    a = 3/0 # try to divide by 0
except ZeroDivisionError as z: # make exception for zerodivisionerror in except block
    print("caught a ZeroDivisionError:",z) # print error
```

```python
# Example 2: OverflowError
import math # need math library to use exponatial in program
try: #try block
    print(math.exp(1000)) # write
except OverflowError as o:
    print("caught an OverflowError:", o)
```

The LookupError class in Python is a base class for exceptions that occur when a key or index is not found. It is the parent class for several specific lookup-related exceptions, including keyerror and indexerror. The purpose of using Lookuperror is to catch any errors related to unsuccessful lookups in collections like dictionaries or sequences.

```python
#KeyError example
dictionary = {'s':19, 'k':11, 'i':9, 'l':12, 'l':12} #create dictionary
try: # try block for error handling
    print(dictionary['r']) #try to print key which does not exist
except KeyError as k: # except block to handle keyerror
    print('Caught a keyerror: ',k) #print error
```

```python
#IndexError example
l = [1,3,5,7] #create a list
try: #try block to handle error
    print(l[6]) #try to print index which is out of range
except IndexError as i: #handle Index error using except
    print('Caught an IndexError:',i) #print error
```
