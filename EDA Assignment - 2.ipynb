**Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.**

```python
import numpy as np
def size_to_kb(size):
    if 'M' in size:
        return float(size[:-1])*1000
    elif 'k' in size:
        return float(size[:-1])
    else:
        return np.nan
df['Size'] = df['Size'].apply(size_to_kb)
```

```python
sns.scatterplot(x=df['Size'],y=df['Rating'])
plt.show()
```

Conclusion: There are apps with large sizes that have high ratings, and there are also apps with large sizes that have low ratings. The same goes for small apps.

**Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.**

```python
df['Type'].unique()
```

```python
df['Price'] = df['Price'].str.replace('$','') # to remove $ sign
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
# Group the DataFrame by app type and calculate the average price for each type
average_prices = df.groupby('Type')['Price'].mean()

# Create a bar chart
average_prices.plot(kind='bar')
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

- Business model: Apps with a subscription-based model may cost less upfront but require ongoing payments, while other apps may have a freemium model where a basic version is free but premium features require payment.
