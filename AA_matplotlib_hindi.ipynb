```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 , 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title ")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c = colours , s = sizes , alpha=.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = ['a' , 'b','c','d','e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
y
```

```python
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python

plt.figure(figsize=(6,2))
plt.plot(x,y,'--b')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title ")
plt.grid()
plt.show()
```

```python
x = [4,5,6,7,8,9]
y = [6,7,8,9,10,11]
plt.plot(x,y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
