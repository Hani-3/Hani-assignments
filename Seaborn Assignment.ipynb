Five plots that we can plot using the Seaborn library and their uses:
- **scatter plot:** A scatter plot is used to visualize the relationship between two variables. It is a great way to see if there is a correlation between two variables, and to identify any outliers.
- **line plot:** A line plot is used to visualize the trend of a variable over time. It is a great way to see how a variable has changed over time, and to identify any patterns or trends.
- **Histogram:** A histogram is used to visualize the distribution of a variable. It is a great way to see how often different values of a variable occur, and to identify any outliers.
- **Bar Plot:** A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. The Bar Chart Displays Data using Several Bars, Each Representing A Particular Category.
- **Pairplot**: Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column. The Diagonal Plots Are A Univariate Distribution Plot That Helps To Draw The Marginal Distribution Of The Data In Each Column. A Pair Plot Pairwise Relationships With Other Columns In The Data Frame And Also Plot Pair Plot With Itself.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri) # make lineplot of timepoint vs signal for different events and regions
```

Question 3

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1) # make subplot to see both plot in one figure this plot at first index
sns.boxplot(x=titanic.pclass, y=titanic.age) # make boxplot of pclass vs age

plt.subplot(1,2,2) # make subplot to see both plot in one figure and this plot at second index
sns.boxplot(x=titanic.pclass, y=titanic.fare) # # make boxplot of pclass vs fare

plt.tight_layout() # to adjust layout
plt.show() # to see plot
```
