**Question 5**

```python
survey = pd.DataFrame({'age':[25,42,37,19,31,28],
                       'soft_drink_Preference':['Coke','Pepsi','Mountain dew','Coke','Pepsi','Coke']
}) ## Given survey data
survey
```

```python
from sklearn.preprocessing import LabelEncoder
## create an instance of label encoder
encoder = LabelEncoder()
```

```python
# Fit and transform the soft drink preference column
encoded = encoder.fit_transform(survey[['soft_drink_Preference']])

# Convert the encoded data to a DataFrame
encoded_survey = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())

# Concatenate the original survey data with the encoded soft drink preference data
survey_encoded = pd.concat([survey, encoded_survey], axis=1)
```

```python
survey_encoded
```

```python
survey_encoded.corr() # to find correlation
```
